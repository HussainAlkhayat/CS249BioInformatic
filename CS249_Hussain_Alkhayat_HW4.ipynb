# Protein Translation Problem

```python
def dict_from_table(data):
    count = 0
    translationTable = {}
    while count < len(data)-1:
        if len(data[count+1]) != 1:
            translationTable[data[count]] = ""
            count += 1
        else:
            translationTable[data[count]] = data[count+1]
            count += 2
    return translationTable
```

```python
def Protein_Translation(rna, translationTable):
    protein = []
    i = 0
    while i < len(rna):
        tempMer = rna[i:i+3]
        protein.append(translationTable[tempMer])
        i += 3
    return protein
```

# Peptide Encoding Problem

```python
def get_rna(dna):    
  # convert string into list
    l = list(dna)  
    for i in range(len(dna)):
        if (l[i] == 'T'):
            l[i] = 'U'
            

    return ''.join(map(str, l))
```

```python
def reverse_dna(dna):
    l = list(dna)  
    for i in range(len(dna)):
        if l[i] == "A":
            l[i] = "T"
        elif l[i] == "C":
            l[i] = "G"
        elif l[i] == "G":
            l[i] = "C"
        elif (l[i] == 'T'):
            l[i] = 'A'
        

    # reverse strand
    l = l[::-1]
    return ''.join(map(str, l))
```

```python
def Peptide_Encoding(peptide, text, translationTable):
    k = len(peptide)*3
    answer = []
    for i in range(len(text) - k + 1):
        tempMer = text[i:i+k]
        #print("tempMer: "+tempMer)
        rna1 = get_rna(tempMer)
        protein1 = Protein_Translation(rna1, translationTable)
        protein1 = ''.join(map(str, protein1))
        rna2 = get_rna(reverse_dna(tempMer))
        protein2 = Protein_Translation(rna2, translationTable)
        protein2 = ''.join(map(str, protein2))
        #print(protein1)
        #print(protein2)
        #print(peptide)
        if protein1 == peptide or protein2 == peptide:
            #print("hi")
            answer.append(tempMer)
    return answer
```

```python
def expand(CandidatePeptides, bases):
    if len(CandidatePeptides) == 0:
        return bases
    CandidatePeptides_copy = []
    for peptide in CandidatePeptides:
        for base in bases:
            CandidatePeptides_copy.append(peptide+base)
    return CandidatePeptides_copy
```

```python
def consistant(peptide, Spectrum):
    if len(peptide) > len(Spectrum):
        return False
    Spectrum_copy = copy.deepcopy(Spectrum)
    for i in peptide:
        if i not in Spectrum_copy:
            return False
        Spectrum_copy.remove(i)
    return True
```

```python
def Generating_Theoretical_Spectrum_(peptide_all, weight):
    weight_all = [0]
    for peptide in peptide_all:
        curr_weight = 0
        for letter in peptide:
            curr_weight += weight[letter]
        weight_all.append(curr_weight)
    return weight_all
```

```python
def generate_all_(peptide):
    gen_all = []
    for i in range(1,int(len(peptide))):
        for j in range(int(len(peptide))-i+1):
            gen_all.append(peptide[j:j+i])
    gen_all.append(peptide)
    return gen_all
```

```python
def CyclopeptideSequencing(Spectrum, weight):
    CandidatePeptides = []
    first = True
    listKeys = weight.keys()
    CandidatePeptides = expand(CandidatePeptides, listKeys)
    goal = Spectrum[-1]
    finalPeptides = []
    count = 0
    while len(CandidatePeptides) != 0:        
        if not first:
            #print(CandidatePeptides)
            print("before: "+ str(len(CandidatePeptides)))
            CandidatePeptides = expand(CandidatePeptides, listKeys)
            print("after: "+ str(len(CandidatePeptides)))
        first = False
        temp = []
        for peptide in CandidatePeptides:
            peptide_all = generate_all_(peptide)
            seq = Generating_Theoretical_Spectrum_(peptide_all,weight)
            #seq = Generating_Theoretical_Spectrum_(peptide,weight)
            con = consistant(seq, Spectrum)
            if seq[-1] == goal:
                if con and peptide not in finalPeptides:
                    finalPeptides.append(peptide)
            elif con:
                temp.append(peptide)
        count+=1
        CandidatePeptides = temp

    return finalPeptides

        
```

# LeaderboardCyclopeptideSequencing

```python
LeaderboardCyclopeptideSequencing(Spectrum, N)
    Leaderboard ← set containing only the empty peptide
    LeaderPeptide ← empty peptide
    while Leaderboard is non-empty
        Leaderboard ← Expand(Leaderboard)
        for each Peptide in Leaderboard
            if Mass(Peptide) = ParentMass(Spectrum)
                if Score(Peptide, Spectrum) > Score(LeaderPeptide, Spectrum)
                    LeaderPeptide ← Peptide
            else if Mass(Peptide) > ParentMass(Spectrum)
                remove Peptide from Leaderboard
        Leaderboard = Trim(Leaderboard, Spectrum, N)
    output LeaderPeptide
```

```python
def LeaderboardCyclopeptideSequencing(Spectrum, weight,N):
    Leaderboard = []
    first = True
    listKeys = list(weight.keys())
    Leaderboard = expand(Leaderboard, listKeys)
    goal = Spectrum[-1]
    count = 0
    LeaderPeptide = ""
    leaderSeq = ''
    while len(Leaderboard) != 0:        
        if not first:
            Leaderboard = expand(Leaderboard, listKeys)
        first = False
        temp = []
        for peptide in Leaderboard:
            peptide_all = generate_all_(peptide)
            seq = Generating_Theoretical_Spectrum_(peptide_all,weight)
            if int(seq[-1]) == int(goal):
                if score_lin(peptide,Spectrum) > score_lin(LeaderPeptide, Spectrum):
                    LeaderPeptide = peptide
                temp.append(peptide)
            elif int(seq[-1]) < int(goal):
                temp.append(peptide)
        Leaderboard = Trim(temp, Spectrum, N)
    return LeaderPeptide

        
```

```python
from operator import itemgetter
def Trim(Leaderboard, Spectrum, N):
    LinearScores = {}
    for peptide in Leaderboard:
        pepScore = score_lin(peptide, Spectrum)
        LinearScores[peptide] = pepScore
    res = dict(sorted(LinearScores.items(), key = itemgetter(1), reverse = True)[:N])
    return res
```

```python
def score_lin(peptide, Spectrum):
    peptide_all = generate_all_(peptide)

    seq = Generating_Theoretical_Spectrum(peptide_all,weight)
    Spectrum_copy = copy.deepcopy(Spectrum)
    count = 0
    for i in seq:
        if str(i) in Spectrum_copy:
            Spectrum_copy.remove(str(i))
            count += 1
    return count
```

```python
def Spectral_Convolution(Spectrum):
    missing = []
    for i in range(len(Spectrum) - 1):
        for j in range(i + 1, len(Spectrum)):
            sub = Spectrum[len(Spectrum)-i-1] - Spectrum[len(Spectrum)-j-1]
            if sub > 0:
                missing.append(sub)
    return missing
```
