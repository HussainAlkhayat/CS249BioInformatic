```python
def addletter(kmers, currStr, len_kmer):
    
    for mer in kmers:
        if currStr[-(len_kmer-1):] == mer[:len(mer)-1]:
            currStr = currStr + mer[-1]
            kmers.remove(mer)
            return kmers, currStr, True
        elif currStr[:len_kmer-1] == mer[1:]:
            currStr = mer[1] + currStr
            kmers.remove(mer)
            return kmers, currStr, True
        
    return kmers, currStr, False
        
    
```

```python
txtFile = "/Users/HussainAlkhayat/Desktop/MS_Fall_2022/CS 249/HW1/ContigGeneration/inputs/input_2.txt"

file = open(txtFile, 'r')

def DeBrujin( data):
    k = len(data[0])
    adjdb = dict()
    for p in data:
        l = p[:k-1]
        r = p[1:]
        if l in adjdb:
            adjdb[l].append(r)
        else:
            adjdb[l] = []
            adjdb[l].append(r)
        if r not in adjdb:
            adjdb[r] = []
    return adjdb


data = list(file.read().strip().split())

adj = DeBrujin(data)
print(adj)
degrees = dict() # [in, out]     
for key in adj.keys():
    degrees[key] = [0,len(adj[key])]

for key in adj.keys():
    for item in adj[key]:
        degrees[item][0] += 1

paths = []
nodes1in1out = set() 
explored = set() 
for v in adj.keys():
    if not (1 == degrees[v][0] and 1 == degrees[v][1]):
        if degrees[v][1] > 0: #out more than 0
            for w in adj[v]:
                currPath = [v, w] 
                while 1 == degrees[w][0] and 1 == degrees[w][1]:
                    explored.add(w)
                    u = adj[w][0]
                    currPath.append(u)
                    w = u
                paths.append(currPath)
    else:
        nodes1in1out.add(v)            
for v in nodes1in1out:
    if v not in explored:
        w = v
        currPath = []
        while w in nodes1in1out:
            currPath.append(w)
            if w == v and len(currPath) > 1:
                paths.append(currPath)
                for node in currPath:
                    explored.add(node)
                break
            w = adj[w][0]
print(paths)
answer = []
for p in paths:
    answer.append(p[0] + ''.join([seq[-1] for seq in p[1:]]))
print(' '.join(answer))
```
