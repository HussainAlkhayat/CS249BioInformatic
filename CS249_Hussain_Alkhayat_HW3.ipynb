```python
def ImmediateNeighbors(Pattern):
    AllLetters = ["A","T","G","C"]
    Neighborhood = []
    for i in range(len(Pattern)):
        letter = Pattern[i]
        for diffLetter in AllLetters:
            if diffLetter != letter:
                tempPattern = list(copy.deepcopy(Pattern))
                tempPattern[i] = diffLetter
                Neighborhood.append("".join(tempPattern))
    return Neighborhood
def NeighborsD(Pattern, d):
    return NeighborsDHelper([Pattern],d)
    
def NeighborsDHelper(Neighbors,d):
    if d == 0:
        return Neighbors
    Ncopy = copy.deepcopy(Neighbors)
    for i in Ncopy:
        temp = ImmediateNeighbors(i)
        for j in temp:
            if j not in Neighbors:
                Neighbors.append(j)
    return NeighborsDHelper(Neighbors,d-1)
```

```python
def MotifEnumeration(DNA, k, d):
    patterns = []
    
    for string in DNA:
        
        for i in range(len(string) - k + 1):
            tempMer = string[i:i+k]
            tempList = NeighborsD(tempMer, d)
            
            for mer in tempList:
                count = 0
                for stringtwo in DNA:
                    for j in range(len(stringtwo) - k + 1):
                        tempMerTwo =  stringtwo[j:j+k]

                        if sum(mer[i] != tempMerTwo[i] for i in range(len(mer))) <= d:
                            #print("mer: "+mer)
                            #print("mertwo: "+tempMerTwo)
                            #print("true")
                            count += 1
                            break
                #print("count: "+ str(count))
                if count == len(DNA):
                    #print("correctmer: "+mer)
                    patterns.append(mer)
    patterns = list(dict.fromkeys(patterns))
    return patterns
```

# MedianString

```python
def hamDis(Text1,Text2):
    dis = 0
    for i in range(len(Text1)):
        if Text1[i] != Text2[i]:
            dis += 1
    return dis
```

```python
def DistanceBetweenPatternAndStrings(Pattern, all_kmers):
    distance = 0
    for string in all_kmers.keys():
        minDis = 1000000000000000
        for mer in all_kmers[string]:
            dis = hamDis(Pattern, mer)
            if dis < minDis:
                minDis = dis
        distance += minDis
    return distance
```

```python
def MedianString(DNA, k):
    all_kmers = {}
    for string in DNA:
        all_kmers[string] = []
        for i in range(len(string) - k + 1):
            tempMer = string[i:i+k]
            if tempMer not in all_kmers[string]:
                all_kmers[string].append(tempMer)
    bases=['A','T','G','C']
    all_possible_kmers = [''.join(p) for p in itertools.product(bases, repeat=k)]
    minDis = 10000000000000000
    kmer = ''
    for mer in all_possible_kmers:
        dis = DistanceBetweenPatternAndStrings(mer,all_kmers)
        if dis < minDis:
            minDis = dis
            kmer = mer
        if dis == minDis:
            #print(dis)
            print(str(dis)+ " mer: "+mer)
            #print("kmer: "+kmer)
    return kmer
        
```

# Profile-most Probable k-mer Problem

```python
def Profile_most_probable(text, k, profile):
    all_kmers = []
    for i in range(len(text) - k + 1):
        tempMer = text[i:i+k]
        if tempMer not in all_kmers:
            all_kmers.append(tempMer)
    minProb = 0
    kmer = ''
    for mer in all_kmers:
        index = 0
        currProb = 1
        for i in range(k):
            currProb *= profile[mer[i]][index]
            index+=1
        if currProb > minProb:
            minProb = currProb
            kmer = mer
    return kmer
```

```python
def mostCommonString(Motifs,k):
    #Find most common Motif
    bases = ["A","C","G","T"]
    mostCommon = []
    for i in range(k):
        count = [0,0,0,0] #A C G T
        for j in range(len(Motifs)):
            letter = Motifs[j][i]
            if letter == "A":
                count[0] += 1
            elif letter == "C":
                count[1] += 1
            elif letter == "G":
                count[2] += 1
            elif letter == "T":
                count[3] += 1
        max_value = max(count) 
        max_index = count.index(max_value)
        mostCommon.append(bases[max_index])
    return mostCommon
```

```python
def score(Motifs, k):
    mostCommon = mostCommonString(Motifs, k)
    currScore = 0
    for i in range(len(Motifs)):
        currScore += hamDis(mostCommon, Motifs[i])
    return currScore
    
```

```python
def createProfile(kmerList):
    k = len(kmerList[0])
    numMer = len(kmerList)
    profile = np.zeros((4, k))
    for mer in kmerList:
        index = 0
        for letter in mer:
            if letter == "A":
                profile[0][index] += 1
            elif letter == "C":
                profile[1][index] += 1
            elif letter == "G":
                profile[2][index] += 1
            elif letter == "T":
                profile[3][index] += 1
            index += 1
    profile = profile / numMer
    return profile
```

```python
def createProfilelap(kmerList):
    k = len(kmerList[0])
    numMer = len(kmerList)
    profile = np.zeros((4, k))
    for mer in kmerList:
        index = 0
        for letter in mer:
            if letter == "A":
                profile[0][index] += 1
            elif letter == "C":
                profile[1][index] += 1
            elif letter == "G":
                profile[2][index] += 1
            elif letter == "T":
                profile[3][index] += 1
            index += 1
    profile = profile + 1
    profile = profile / (numMer+4)
    return profile
```

# RandomizedMotifSearch

```python
def Profile_most_probable_rand(text, k, profile):
    all_kmers = []
    for i in range(len(text) - k + 1):
        tempMer = text[i:i+k]
        if tempMer not in all_kmers:
            all_kmers.append(tempMer)
    minProb = 0
    kmer = ''
    for mer in all_kmers:
        index = 0
        currProb = 1.0
        for letter in mer:
            if letter == "A":
                currProb *= profile[0][index]
            elif letter == "C":
                currProb *= profile[1][index]
            elif letter == "G":
                currProb *= profile[2][index]
            elif letter == "T":
                currProb *= profile[3][index]
            index+=1
        if currProb > minProb:
            minProb = currProb
            kmer = mer
    return kmer
```

```python
def RandomizedMotifSearch(Dna, k, t):
    allKmers = []
    for text in Dna:
        tempList = []
        for i in range(len(text) - k + 1):
            tempList.append(text[i:i+k])
        allKmers.append(tempList)
    BestMotifs = []
    tempMotifs = []
    for i in range(len(Dna)):
        index = random.randint(0, len(allKmers[i])-1)
        tempMotifs.append(allKmers[i][index])
    BestMotifScore = 10000000000
    while True:
        profile = createProfilelap(tempMotifs)
        tempMotifs = []
        for i in range(t):
            tempMotifs.append(Profile_most_probable_rand(Dna[i], k, profile))
        
        tempMotifScore = score(tempMotifs, k)
        if tempMotifScore < BestMotifScore:
            BestMotifScore = tempMotifScore
            BestMotifs = tempMotifs
        else:
            return BestMotifs
    
    return BestMotifs
    
```

```python
def temofunc(Dna, k, t):
    BestMotifScore = 10000000000
    BestMotifs = []
    for i in range(1000):
        print(i)
        tempMotifs = RandomizedMotifSearch(Dna, k, t)
        tempMotifScore = score(tempMotifs, k)
        if tempMotifScore < BestMotifScore:
            BestMotifs = tempMotifs
            BestMotifScore = tempMotifScore
    return BestMotifs
```

# GibbsSampler

```python
GibbsSampler(Dna, k, t, N)
    randomly select k-mers Motifs = (Motif1, …, Motift) in each string from Dna
    BestMotifs ← Motifs
    for j ← 1 to N
        i ← Random(t)
        Profile ← profile matrix constructed from all strings in Motifs except for Motifi
        Motifi ← Profile-randomly generated k-mer in the i-th sequence
        if Score(Motifs) < Score(BestMotifs)
            BestMotifs ← Motifs
    return BestMotifs
```

```python
def profile_randomized_kmer(dna, k, prof):
    nuc_loc = {nucleotide: index for index, nucleotide in enumerate('ACGT')}
    probs = []
    for i in range(len(dna) - k):
        current_prob = 1.
        for j, nucleotide in enumerate(dna[i:i + k]):
            current_prob *= prof[nuc_loc[nucleotide]][j]
        probs.append(current_prob)

    i = np.random.choice(len(probs), p = np.array(probs) / np.sum(probs))
    return dna[i:i + k]
```

```python
def GibbsSampler(Dna, k, t, N):
    allKmers = []
    for text in Dna:
        tempList = []
        for i in range(len(text) - k + 1):
            tempList.append(text[i:i+k])
        allKmers.append(tempList)
    BestMotifs = []
    tempMotifs = []
    for i in range(len(Dna)):
        index = random.randint(0, len(allKmers[i])-1)
        tempMotifs.append(allKmers[i][index])
    BestMotifScore = score(tempMotifs, k)
    for i in range(N):
        j = random.randint(0,len(tempMotifs)-1)
        profile = createProfilelap(tempMotifs[:j] + tempMotifs[j+1:])
        tempMotifs[j] = profile_randomized_kmer(Dna[j], k, profile)
        tempMotifScore = score(tempMotifs, k)
        if tempMotifScore < BestMotifScore:
            BestMotifScore = tempMotifScore
            BestMotifs = tempMotifs
    return BestMotifs
```

```python
def temofunc2(Dna, k, t,N):
    BestMotifScore = 10000000000
    BestMotifs = []
    for i in range(21):
        tempMotifs = GibbsSampler(Dna, k, t,N)
        tempMotifScore = score(tempMotifs, k)
        if tempMotifScore < BestMotifScore:
            BestMotifs = tempMotifs
            BestMotifScore = tempMotifScore
    return BestMotifs
```

```python
def profile_with_pseudocounts(motifs):
    prof = []
    for i in range(len(motifs[0])):
        col = ''.join([motifs[j][i] for j in range(len(motifs))])
        prof.append([float(col.count(nuc)+1)/float(len(col)+4) for nuc in 'ACGT'])
    return prof
def score(motifs):
    score = 0
    for i in range(len(motifs[0])):
        motif = ''.join([motifs[j][i] for j in range(len(motifs))])
        score += min([hamming_distance(motif, homogeneous*len(motif)) for homogeneous in 'ACGT'])
    return score
def hamming_distance(str1, str2):
    counter = 0
    for s1, s2 in zip(str1, str2):
        if s1 != s2:
            counter += 1
    return counter


def profile_randomized_kmer(dna, k, prof):
    nuc_loc = {nucleotide: index for index, nucleotide in enumerate('ACGT')}
    probs = []
    for i in range(len(dna) - k):
        current_prob = 1.
        for j, nucleotide in enumerate(dna[i:i + k]):
            current_prob *= prof[j][nuc_loc[nucleotide]]
        probs.append(current_prob)

    i = np.random.choice(len(probs), p = np.array(probs) / np.sum(probs))
    return dna[i:i + k]


def gibbs_sampling_motif_search(dna_list, k, t, N, init_motifs=None):
    if init_motifs:
        motifs = init_motifs
    else:
        rand_ints = [random.randint(0, len(dna_list[0]) - k) for a in range(t)]
        motifs = [dna_list[i][r:r + k] for i, r in enumerate(rand_ints)]

    best_score = [score(motifs), list(motifs)]

    for j in range(N):
        i = random.randint(0, t - 1)
        current_profile = profile_with_pseudocounts([x for amotif, x in enumerate(motifs) if amotif != i])
        motifs[i] = profile_randomized_kmer(dna_list[i], k, current_profile)
        current_score = score(motifs)
        if current_score < best_score[0]:
            best_score = [current_score, list(motifs)]

    return best_score
```
